## Exploratory Data Analysis (EDA)

Question:
1. Berapa persen proses pengiriman yang tidak sesuai dengan estimasi pengiriman?
2. Proses apa yang paling berpengaruh pada keterlambatan pengiriman?

```python
def get_time_difference(start_time, end_time, unit='seconds', format="%Y-%m-%d %H:%M:%S"):
  start_time = datetime.strptime(start_time, format)
  end_time = datetime.strptime(end_time, format)
  difference = end_time - start_time
  if unit == 'seconds':
    return difference.total_seconds()
  elif unit == 'minutes':
    return difference.total_seconds() / 60
  elif unit == 'hours':
    return difference.total_seconds() / 3600
  elif unit == 'days':
    return difference.days + difference.seconds / 86400
  else:
    raise ValueError("Unit must be one of 'seconds', 'minutes', 'hours', or 'days'")

def get_is_delivery_late(delivered_time, estimated_time):
  diff = get_time_difference(estimated_time, delivered_time)
  return diff > 0
```

### Explore the number of late deliveries

```python
orders_data["is_delivery_late"] = orders_data.apply(
    lambda row: get_is_delivery_late(row['order_delivered_customer_date'], row['order_estimated_delivery_date']),
    axis=1
)
```

```python
late_orders_data = orders_data[orders_data['is_delivery_late'] == True]
```

```python
late_orders_data.describe(include='all')
```

### Explore the duration for each delivery process

```python
late_orders_data['delivery_late_time'] = orders_data.apply(
    lambda row: get_time_difference(row['order_estimated_delivery_date'], row['order_delivered_customer_date'], 'hours'),
    axis=1
)
```

```python
late_orders_data[["delivery_late_time"]].describe()
```

```python
late_orders_data["purchase_to_approve_time"] = late_orders_data.apply(
    lambda row: get_time_difference(row['order_purchase_timestamp'], row['order_approved_at']),
    axis=1
)
late_orders_data["approve_to_delivered_carrier"] = late_orders_data.apply(
    lambda row: get_time_difference(row['order_approved_at'], row['order_delivered_carrier_date']),
    axis=1
)
late_orders_data["delivered_carrier_to_delivered_customer"] = late_orders_data.apply(
    lambda row: get_time_difference(row['order_delivered_carrier_date'], row['order_delivered_customer_date']),
    axis=1
)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Berapa persen proses pengiriman yang tidak sesuai dengan estimasi pengiriman?

```python
late_counts = orders_data['is_delivery_late'].value_counts()
labels = ['On Time', 'Late']
sizes = late_counts

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%',)
plt.title('Proportion of On-Time vs Late Deliveries')
plt.show()
```

### Pertanyaan 2: Proses apa yang paling berpengaruh pada keterlambatan pengiriman?

```python
correlations = late_orders_data[['purchase_to_approve_time', 'approve_to_delivered_carrier', 'delivered_carrier_to_delivered_customer', 'delivery_late_time']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

## Conclusion

- Ada 8.1% pengiriman yang mengalami keterlambatan dari waktu estimasi
- Proses yang paling berpengaruh terhadap keterlambatan pengiriman adalah proses pengantaran dari ekspedisi pengiriman ke customer
